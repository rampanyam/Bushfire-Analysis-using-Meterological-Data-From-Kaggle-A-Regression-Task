## ii) Handling skewed data  (Variable Transformations):

#### Log and Square root Transformations:

    Log transformation and sqiare root transformations are applied to handle skewed data.
    
    Log transformation is used to make highly skewed data less skewed and often helps in the interpretation task and prediction task.
    
    The Square root transformation has a moderate effect on the skeweness as opposed to the log transformation. 


    i) Wind has no zeroes, hence both square root and log(x+1) transformation can be performed.

    ii) Rain has 98 percent zeroes, hence, log(x+1) transofrmation can be performed.(Adding a Constant). 

    iii) RH has no zeroes, hence both square root and log(x+1) transofrmation can be performed.

    iv) DMC has no zeroes, both square root and log(x+1) transformation can be performed.

    v) ISI has 1 zero, hance a constant is added while transformation, log(x+1) transoformation is performed. 
    
    vi) Area has 47 % zeroes, hence, log(x+1) transformation should be performed on area. As it is the Dependent Variable
        and as it is  highly skewed a log(x+1) transformation should be used. 
    
    vii) FFMC  and DC have no zero , hence both log and square root transformation can be applied.

```r
# the log transformations are computed using the log() function
# The histograms are plotted for each transformed variable using ggplot()
# The geom_hist() function is used to plot histogram. 

# The sqrt() function is used to compute the square root transformation. 
# 1.) Wind:
#log transformation for wind: 
wind_log <- ggplot(forestfires,aes(x=log((wind)+1)))+geom_histogram(bins=8,fill = 'lightgreen',color = 'black')+ggtitle("log(Wind)")
# square root :
wind_squareroot <- ggplot(forestfires,aes(x=sqrt(wind)))+geom_histogram(bins=8,fill = 'darkgreen',color = 'black')+ggtitle("sqrt(Wind)")

#  2 ) Rain : 
# log (x + 1)
rain2 <- ggplot(forestfires,aes(x=log(rain+1)))+geom_histogram(bins=10,fill = 'red',color = 'black')+ggtitle("log(rain+1)")


# 3)  Relative Humidity : 
### Log transformation:
relative_humidity_log<-ggplot(forestfires,aes(x=log(RH+1)))+geom_histogram(bins=8,fill = 'red1',color = 'black')+ggtitle("log(RH)")
### Square root transformation:
relative_humidity_sqrt<-ggplot(forestfires,aes(x=sqrt(RH)))+geom_histogram(bins=8,fill = 'orange',color = 'black')+ggtitle("sqrt(RH)")



# 4 ) Dc Square root and log 
DClog <- ggplot(forestfires,aes(x=log(DC+1)))+geom_histogram(bins=10,fill = 'brown',color = 'black')+ggtitle("log(DC)")
DCsquareroot <- ggplot(forestfires,aes(x=sqrt(DC)))+geom_histogram(bins=10,fill = 'pink',color = 'black')+ggtitle("sqrt(DC)")



#5 ) FMMC Square root and log transformation:
FFMClog <- ggplot(forestfires,aes(x=log(FFMC+1)))+geom_histogram(bins=10,fill = 'grey',color = 'black')+ggtitle("log(FFMC)")
FFMCsquareroot <- ggplot(forestfires,aes(x=sqrt(FFMC)))+geom_histogram(bins=10,fill = 'skyblue',color = 'black')+ggtitle("sqrt(FFMC)")


# ISI log transformation:
ISIlog <- ggplot(forestfires,aes(x=log(ISI+1)))+geom_histogram(bins=10,fill = 'white',color = 'black')+ggtitle("log(ISI+1)")


# DMC log and square root:
DMClog <- ggplot(forestfires,aes(x=log(DMC+1)))+geom_histogram(bins=10,fill = 'steelblue',color = 'black')+ggtitle("log(DMC)")
DMCsquareroot <- ggplot(forestfires,aes(x=sqrt(DMC)))+geom_histogram(bins=10,fill = 'red4',color = 'black')+ggtitle("sqrt(DMC)")



```

## 3. Model Development<a class="anchor" id="sec_3"></a>

### 1.) Feature Selection:


### Manual Selection : 
Applying Linear Regression to identify significant variables based on P - Values and taking into consideration correlations and interaction.

### General Model with the response variables and all the variables:

```r
generalmodel <- lm(area ~ . , final_forest_fires)
summary(generalmodel) # Training a linear model using the lm() function the ~ . represents all the variables other than the response variable 
```

Negative Adjusted R2  indicates that the residual sum of squares is almost equal to the total sum of squares.

Hence, it indicates that  the explanation towards response is very very low or negligible. 

So, Negative Adjusted R2 means that the  explanatory variables are indsignificant. 

Moreover, as the area is skewed and has over 47% zeroes, a log transformation would yeild better results.


The P - Value is greater than the significance Level and hence we fail to reject null hypothesis. ( Model does not fit the data well). 

### Model based on High Correlation with area and P - Values 

Hence from the baseline model and the explorations above,  the variables 

temp, DMC, DC, X, monthaug, monthjul, monthjun, monthmar, monthapr are considered. 

```r
model_one <- lm(area ~ X+DMC+DC+temp+monthaug+monthjul+monthjun+monthmar+monthapr,final_forest_fires)
summary(model_one) # Training a linear model using the lm() function the syntax is response ~ predictors.
# The summary of the linear regression returns the metrics 
```

The adjusted R Squared of the Model is 0.0094, it indicates that the model explains 0.9% varaibality of the response variable around its mean. This model is better than the previous model in terms of Adjusted R-Squared, but the P - Value is greater than alpha , hence we fail to  reject the Null Hypotheis, there is no linearity between X and Y. Moreover, the F- Statistic is greater  than the previous model and it indicates slightly higher significance . 

## Adding Interactions between highly correlated variables to the previous model : 

Interactions between two variables A and B are added to check whether  A and B together impact the dependent/ target variable or not. 

The variables having high correlation are chosen in this case.

1.) DMC and DC 

2.) ISI and FFMC  

3) DMC and Temperature 

4) X and Y 



```r
model_two <- lm(area ~ X+DMC+DC+temp+monthaug+monthjul+monthjun+monthmar+monthapr+X:Y+DMC:DC+ISI:FFMC+X:Y,final_forest_fires)
summary(model_two) # Training a linear model using the lm() function the y~ x. 
```

The Adjusted R-Squared Increased to 0.013, i.e. the model can explain 1.3% of Variability of the dependent variable
around its mean. However, there is very minute improvement in the Residual Standard Error. 
There is a very little increase in the F-Statistic. This indicates that the model has some sort of significance.
Moreoverl, the P=Value is greater than alpha, hencd, we fail to reject the null hypothesis. Hence, there is no linearity between X and y.

Hence, as the dependant variable is highly skewed , a log(area+1) transformation on the variable would be ideal.

### PLOTTING INFLUENTIAL POINTS 

### What are influential points ? 

Influential points are the outlier points that affect the slope of the regression line. The influential points may cause the coeffecient of determination to be bigger. Hence, the regression is computed with and without the influential points. 
Cooks Distance is used to analyse the influence of a data point while performing regression.  
The points whicha are greater than 4 times the mean of cooks distance is  chosen as the regression line. 



```r
# Cooks Distance for Influential Points
# Plotting 
cooksd <- cooks.distance(model_four_log)
plot(cooksd, pch="*", cex=2, main="Influential Obs by Cooks distance")  # plot cook's distance
abline(h = 4*mean(cooksd, na.rm=T), col="red")  # add cutoff line
text(x=1:length(cooksd)+1, y=cooksd+0.01, labels=ifelse(cooksd>4*mean(cooksd, na.rm=T),names(cooksd),""), col="red")  # add labels
```
